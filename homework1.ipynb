#### *На языке Python предложить алгоритм вычисляющий численно предел с точностью E = 10 в степени -7

```python
a = [0, 1, 1]
for n in range(3, 15):
    x = n / ((math.factorial(n))**(n/2))
    a.append(x)
    r = abs(x - a[n-1])
    if r < 10**(-7):
        print(r)
        break
```

```python
a
```

```python
print('Предел, удовлетворяющий точности:', a[-1])
```

#### Предложить оптимизацию алгоритма, полученного в задании 3, ускоряющую его сходимость.

Для оптимизации увеличиваем шаг

```python
b = [0, 1, 1]
n = 3
while True:
    x = n / ((math.factorial(n))**(n/2))
    r = abs(x - b[-1])
    b.append(x)
    if r < 10**(-7):
        print(r)
        break
    n += 2
```

```python
b
```

```python
print('Предел, удовлетворяющий точности:', b[-1])
```
